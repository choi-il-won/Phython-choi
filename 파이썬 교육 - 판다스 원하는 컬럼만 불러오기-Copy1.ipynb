# 일부 컬럼을 기준으로 데이터 가져오기 

```python
import pandas as pd 
import seaborn as sns 
```

```python
df = sns.load_dataset("iris")
df
```

```python
#원하는 컬럼 불러오기 
columns = ['sepal_width','sepal_length','species']
df[columns].head()
```

```python
#특정 컬럼만 가져오기 
#[]를 사용해서 가져오기
df['sepal_width']
```

```python
#regex 정규표현식 
df.filter(regex ="_")

#특정 문자만 가져오고 싶을 때 (^ 필수 )
df.filter(regex ="^sepal")

#특정 문자만 가져오고 싶을 때 = 특정 문자가 끝날 때는 (문자 뒤에 $)
df.filter(regex = "es$")

#특정 문자만 빼고 가져오고 싶을 때 = [^(species$).*] , *가 (?!..$)안에 있는 문자 빼고 가져와라
df. filter(regex = '^(?!species).*')
```

```python
#X부터 X4까지 컬럼을 한 번에 가져오고 싶을떄는 loc쓰고 : 사용 

df.loc [:,"sepal_width": 'petal_width']

# 컬럼 번호로 써도 가져올 수 있음 

df.loc [2:5]
```

```python
# : 모든 행을 불러오고 , [1,2,4]는 지정된 열만 불러오기 
# iloc는 인덱스 전 번호까지 가져옴 
# loc는 인덱스 번호까지 가져옴 
#df.iloc[:, [1,2,4]]

#뒷번호 부터 보고 싶으면 - 사용 
df.iloc[-5 :,[1,3]]
```

```python
# 조건에 맞는 컬럼 불러오기 
#'sepal_width' , 'petal_length' 컬럼에서 "sepal_length" 값이 3보다 큰 것만 불러오기 

df.loc[df["sepal_length"]>3 ,['sepal_width' , 'petal_length']]
```
