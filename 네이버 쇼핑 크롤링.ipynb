```python
import requests 
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://search.shopping.naver.com/search/all?query=%EC%BB%A4%ED%94%BC%EB%A8%B8%EC%8B%A0&frm=NVSHATC&prevQuery=%EC%BB%A4%ED%94%BC%EB%A8%B8%EC%8B%A0'
url
```

```python
res =requests.get(url)
res
```

```python
type(res.text)
```

```python
html = bs(res.text)
html
```

## 가격 데이터 긁어오기 

```python
prices =html.select('.price_num__2WUXn')
prices
```

## List comprehension 사용 하여 가격 추출하기

```python
coffee_prices =[price.text for price in prices]
coffee_prices
```

## 커피머신 데이터, 링크 긁어오기

```python
names =html.select(".basicList_link__1MaTN  ")
names
```

## List comprehension 사용하여 커피머신 이름 추출하기

```python
coffee_names = [title.text for title in names]
coffee_names
```

## List comprehension 사용하여 커피머신 판매 링크 추출하기

```python
coffee_links = [title["href"] for title in names]
coffee_links
```

```python
df_coffee = pd.DataFrame({
    "name" : coffee_names
    "price" : coffee_prices
    "link": coffee_links
}) 
df_coffee
```
